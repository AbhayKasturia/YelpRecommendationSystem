```python
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from collections import Counter
import operator
%matplotlib inline
TOP = 15
import json
import matplotlib.pyplot as plt
```

```python
growth_data = {}

with open('../../dataset/user.json') as data_file:
    for line in data_file:        
        data = json.loads(line)        
        if data['yelping_since'][0:4] in growth_data:
            growth_data[data['yelping_since'][0:4]] += 1
        else:
            growth_data[data['yelping_since'][0:4]] = 1
```

```python
ts = Series(list(growth_data.values()), index=list(growth_data.keys()))
ts = ts.cumsum()
ts = ts.plot()
ts.set_xlabel('Year')
ts.set_ylabel('Count of Yelp Users')
```
